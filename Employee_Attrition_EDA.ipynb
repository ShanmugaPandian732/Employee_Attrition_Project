```python
# Import the Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the Dataset

df = pd.read_csv(r"C:\Users\arun prakash\Downloads\Employee-Attrition - Employee-Attrition.csv")
df
```

### Distribution of Employee Attrition

```python
# Calculate the value counts of the 'Attrition' column
attrition_counts = df['Attrition'].value_counts()

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Employee Attrition')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Attrition with Respect to Business Travel

```python
# Group by BusinessTravel and Attrition and count the occurrences
attrition_by_businesstravel = df.groupby(['BusinessTravel', 'Attrition']).size().unstack()

# Create a stacked bar plot
attrition_by_businesstravel.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Attrition with Respect to Business Travel')
plt.xlabel('Business Travel')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Age Distribution of Employees

```python
# Calculate the frequency of each age
age_counts = df['Age'].value_counts().sort_index()

# Create a line plot of age distribution
plt.figure(figsize=(10, 6))
plt.plot(age_counts.index, age_counts.values)
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.grid(True)
plt.show()
```

### Monthly Income Distribution of Employees

```python
# Create bins for Monthly Income
income_bins = pd.cut(df['MonthlyIncome'], bins=10)  # You can adjust the number of bins

# Calculate the frequency of each bin
income_bin_counts = income_bins.value_counts().sort_index()

# Create a bar plot of monthly income distribution by bin
plt.figure(figsize=(12, 6))
income_bin_counts.plot(kind='bar')
plt.title('Monthly Income Distribution of Employees')
plt.xlabel('Monthly Income')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Attrition with Respect to Monthly Income

```python
# Create a box plot of MonthlyIncome with respect to Attrition
plt.figure(figsize=(8, 6))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title('Attrition with Respect to Monthly Income')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.tight_layout()
plt.show()
```

### Distribution of JobSatisfaction, EnvironmentSatisfaction, RelationshipSatisfaction and JobInvolvement by Employees


```python
# Create a figure with subplots for the four pie charts
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()  # Flatten the 2x2 array of axes for easy iteration

columns_to_plot = ['JobSatisfaction', 'EnvironmentSatisfaction', 'RelationshipSatisfaction', 'JobInvolvement']

for i, col in enumerate(columns_to_plot):
    counts = df[col].value_counts()
    axes[i].pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
    axes[i].set_title(f'Distribution of {col}')
    axes[i].axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.

plt.tight_layout()
plt.show()
```
